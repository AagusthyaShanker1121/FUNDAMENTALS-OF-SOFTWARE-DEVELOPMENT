### Exercise 1: Write a program to initaliaze two variables x to 4 and y to 2. then print out the results of the following operations (format each result to 3 decimal points and print each result on a new line) 

```python
import numpy as np
x = 4
y = 2
```

```python
print(np.round(x + y, 3))
print(np.round(x - y, 3))
print(np.round(x/y, 3))
print(np.round(x*y, 3))
print(np.round(x%y + x/y, 3))
result = (y**7 + 7/(np.sqrt(5)+x)) * (x**4 % 5 + 2)
print(np.round(result, 3))
```

### Exercise 2: Write a program that completes the following operations:
1. Read in the integer values of x and y from STDIN
2. Declare a variable z, and assign the result of x to the power of y to it
3. Declare a variable called 'result' and assign the square-root of z to it
4. print the above results each on a new line (formatted to 2 decimal points)

```python
x = int(input("Enter the value of x:\t"))
y = int(input("Enter the value of y:\t"))
z = np.round(x**y, 2)
result = np.round(np.sqrt(z), 2)
print(f'x = {x}')
print(f'y = {y}')
print(f'z = {z}')
print(f'result = {result}')
```

# Exercise 3: Write a program that outputs the following and determine the area and volume
1. Enter circle radius: xyz
2. Circle area of radius "value of radius go here" is : area
3. Sphere volume of radius "value of radius go here" is: volume


```python
radius = float(input("Enter the radius of the circle:\t"))
area = np.pi * radius**2
sphere_volume = 4/3 * np.pi * radius**3
print(f'Circle area of radius "{radius}" units is: {np.round(area, 2)} square units')
print(f'Sphere volume of radius "{radius}" is: {np.round(sphere_volume, 2)} cube units')
```

```python
x1 = float(input("Enter the coordinate value of x1 of point A:\t"))
y1 = float(input("Enter the coordinate value of y1 of point A:\t"))
x2 = float(input("Enter the coordinate value of x2 of point B:\t"))
y2 = float(input("Enter the coordinate value of y2 of point B:\t"))

coordinate_distance = np.round(
    np.sqrt(
        (x1 - x2)**2 + (y1 - y2)**2
    ),
    2
)

print(f'The distance between A(x1, y1) and B(x2, y2) is distance {coordinate_distance} units')
```

```python
class Car():
    def __init__(self, make:str, pos:float):
        self.make = make
        self.pos = pos
    
    def move(self, distance:float):
        self.pos += distance
    
    def __str__(self):
        return f'{self.make} is at position {self.pos}'

bmw = Car("BMW", 0)
print(bmw)
print(f'BMW is at position {bmw.pos}')
bmw.move(15)
print(f'BMW is at position {bmw.pos}')

```

```python

class Numbers():
    def __init__(self, arr):
        self.arr = arr
        self.size = len(arr)
    def __str__(self):
        return str(arr)
    def change_element(self, index:int, new_number:int):
        self.arr[index] = new_number


size = int(input("Enter the size of array:\t"))
arr = []
for i in range(size):
    num = int(input(f"Enter num:\t"))
    arr.append(num)

number = Numbers(arr)
print(number)
number.change_element(0, 10)
number.change_element(-1, -5)
mid_idx = lambda: number.size // 2 - 1 if number.size % 2 == 0 else number.size // 2
print(mid_idx())
number.change_element(mid_idx(), 3)
print(number)
        
```
