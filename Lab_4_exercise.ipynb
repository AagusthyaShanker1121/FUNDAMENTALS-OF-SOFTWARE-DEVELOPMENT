```python
while True:    
    n = int(input("Enter the number of elements 'n':\nPress -1 to quit the execution"))
    if n == -1:
        break
    if n > 1:
        sum = 0
        for i in range(2, n + 1, 2):
            sum += i
        print(sum)
    else:
        print('Invalid n')
        
```

```python
total_vowels = 0
final_character = ''
while True:
    character = input("Enter character:\t").lower()
    if character == '.':
        break
    if '.' in character:
        character = character[:character.find('.') - 1]
    final_character += character
for i in final_character:
    if i in ['a', 'e', 'i', 'o', 'u']:
        total_vowels += 1

print(final_character)
print(total_vowels)
```

```python
n = int(input("Enter n:\t"))
fact = 1
for i in range(1, n+1):
    fact *= i
print(fact)
```
