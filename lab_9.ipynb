```python
def frequency(user_input: str) -> dict:
    vowels = ['a', 'A', 'e', 'E', 'o', 'O', 'i', 'I', 'u', 'U']
    vowel_frequency = {vowel:0 for vowel in vowels}

    for letter in user_input:
        if letter in vowels:
            vowel_frequency[letter] += 1
    
    return vowel_frequency

def preprocess_text(vowel_frequency) -> str:
    text = ""
    for key, value in vowel_frequency.items():
        text += f"{key} --> {value}\n"
    
    return text[:-1]

def fileOperations(strings: list) -> None:
    with open(file = 'vowel.txt', mode = '+a') as vowel_file:
        final_text = ""
        for string in strings:
            vowel_frequency_text = preprocess_text(frequency(string))
            final_text = final_text + f"String: {string}\n Vowel Frequencies:\n {vowel_frequency_text}\n"
        
        vowel_file.write(final_text)
        vowel_file.seek(0)
        print(vowel_file.read())

def main():
    # Taking user inputs until user inputs *
    user_input = input("Enter string:\t")
    strings = []
    while user_input != '*':
        strings.append(user_input)
        user_input = input("Enter string:\t")
    
    # Doing the main task using fileOperation function. writing and reading vowel frequencies of each text.
    fileOperations(strings)

# Running the main file
main()
```
